A person throws a ball upwards at a speed of 10 m/s, please calculate the trajectory using Runge-Kutta method and plot height (z) vs time (t).

```python
g = 9.81
h = 0.01
t = 0
y = 0

def f(t):
    return 10-g*t

def rk4(t, h, y, n):
    k1 = np.zeros(n)
    k2 = np.zeros(n)
    k3 = np.zeros(n)
    k4 = np.zeros(n)
    k1=h*f(t)
    k2=h*f(t+0.5*h)
    k3=h*f(t+0.5*h)
    k4=h*f(t+h)
    y+=(k1 + 2*k2 + 2*k3 + k4)/6
    return y

zp=[]
tp=[]
tend = 2.5
while(t<tend):
    if((t+h)>tend):
        h=tend-t
    y = rk4(t,h,y,2)
    zp=np.append(zp,y)
    tp=np.append(tp,t)
    t = t+h
    
# compate numerical solution with analytic solution 
def an_z(t):
    return 10*t - 0.5*g*t**2

plt.plot(tp,zp, label = 'rk4')
plt.plot(tp, an_z(tp), label = 'analytic')
plt.legend()
plt.show()
```
