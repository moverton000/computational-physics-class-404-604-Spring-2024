```python
import matplotlib.pyplot as plt
import numpy as np

#x = np.array([-1, -0.5, 0.5, 1])
x = np.arange(-np.pi, np.pi+0.05, 0.05)

#func = (np.pi/2) + 2 * ( np.sin(3*x) / 3 ) + np.sin(5*x) / 5 

def func(n, x): 
    # n is the max index and x is an array 
    
    fval = np.pi/2
    
    for i in range(1, n+1, 2): 
        fval = fval + 2*np.sin(i*x)/i 
        
    return fval

plt.figure(figsize=(10, 8))

plt.plot(x/np.pi, func(11, x))
plt.plot(x/np.pi, func(21, x))
plt.plot(x/np.pi, func(31, x))

plt.xlabel('x values', fontsize=20)
plt.xticks([-1,-0.75,-0.5,-0.25, 0, 0.25, 0.5, 0.75, 1],['-$\\pi$', '-$\\frac{3\\pi}{4}$','-$\\frac{\\pi}{2}$', '-$\\frac{\\pi}{4}$', '0', '$\\frac{\\pi}{4}$', '$\\frac{\\pi}{2}$', '$\\frac{3\\pi}{4}$', '$\\pi$'], fontsize=20)
plt.ylabel('f(x)', fontsize=20)
plt.grid()
```
